# Exploratory Data Analysis

## Q.1.Find how many movies & TV shows released in the year 1996?Show in bar Graph.


```python
df[df['Release_year'] == 1996].groupby('Release_year')['Rating'].count().plot(kind='bar')
```

## Q.2.For 'Angry Birds', what is the rating,duration and release date ?

```python
#to check particular record
df[df['Title'].isin(["Angry Birds"])]
```

```python
df[df['Title'].str.contains("Angry Birds")]
```

## Q.3. How many Movies & TV Shows are in the dataset ? Show with Piechart.

```python
#Grouping Unique Item and Counting their values
df.groupby('Type').Type.count()
```

```python
#Representing the grouped items in piechart
plt.figure(figsize=(12,6))
plt.title("Percentage of Movies or TV Shows")
g = plt.pie(df.type.value_counts(),explode=(0.025,0.025),labels=df.type.value_counts().index, colors=['pink','skyblue'],autopct='%1.1f%%', startangle=180)
plt.show()
```

## Q.4.What are the different Ratings defined by Netflix ?Show with Bar Graph

```python
order =  ['G', 'TV-Y', 'TV-G', 'PG', 'TV-Y7', 'TV-Y7-FV', 'TV-PG', 'PG-13', 'TV-14', 'R', 'NC-17', 'TV-MA']
plt.figure(figsize=(15,7))
g = sns.countplot(df.rating, hue=df.type, order=order);
plt.title("Ratings for Movies & TV Shows")
plt.xlabel("Rating")
plt.ylabel("Total Count")
plt.show()
```
